For simplicity we reduced the original 106 predictors down to columns that we think will have a noticeable effect on hover or not hover along with the turns. This will reduce the computational requirements for some of the machine learning models. 

```python
# initalize hover or not, left turn, right turn column
df2 = df[['latitude',
          'longitude', 
          'absoluteAltitude', 
          'trueHeading', 
          'groundSpeed', 
          'trueAirspeed', 
          'climbOrDescentRate', 
          'hasWeightOnWheels', 
          'turnRate to degPerSec', 
          'cyclicPositionPitch', 
          'cyclicPositionRoll', 
          'collectivePosition', 
          'antiTorquePedalPosition', 
          'throttlePosition', 
          'e1Torque', 
          'e2Torque',
]]

# Create dummy variable columns related to the maneuvers. These columns become 0 or 1 based off the flight logs. 
df2['HoverOrNot'] = 0
df2['RightTurn'] = 0
df2['LeftTurn'] = 0

df2
```

For this I calculated several statistics and assigned them a new 

```python
df2['turnDirection'] = df2['trueHeading'].rolling(window = '5s').apply(lambda x: x[-1]-x[0], raw=True)
df2['Altdiff'] = df2['absoluteAltitude'].rolling(window = '20s').apply(lambda x: x[0]- x[-1], raw = True)
df2['longDiff'] = df2['longitude'].rolling(window = '20s').apply(lambda x: x[-1]-x[0], raw=True)
df2['latDiff'] = df2['latitude'].rolling(window = '20s').apply(lambda x: x[-1]-x[0], raw=True)
```

```python
df2
```

```python
def hover(x): 
    if x['absoluteAltitude'] > 2 and x['absoluteAltitude'] < 50 and x['groundSpeed'] < 2 and x['Altdiff'] < 10: return 10
    else: return 0 
df2['LabHoverOrNot'] = df2.apply(hover, axis = 1)

def LTurn(x): 
    if x['LabHoverOrNot'] == 15 and x['turnDirection'] < -10: return 5
    else: return 0

def RTurn(x): 
    if x['LabHoverOrNot'] == 15 and x['turnDirection'] > 10: return 7
    else: return 0 

df2['LabRightTurn'] = df2.apply(RTurn, axis = 1)
df2['LabLeftTurn'] = df2.apply(LTurn, axis = 1)
```

```python

df3 = df2
```

```python
# Assign Hover Or Not based on the log
df3.loc[(df3.index >= '2022-06-09 16:00:24') & (df3.index <= '2022-06-09 16:06:46'),['HoverOrNot']] = 10
# Assign right turn based on the log
df3.loc[(df3.index >= '2022-06-09 16:00:24') & (df3.index <= '2022-06-09 16:00:56'),['RightTurn']] = 7
df3.loc[(df3.index >= '2022-06-09 16:01:28') & (df3.index <= '2022-06-09 16:01:57'),['RightTurn']] = 7
df3.loc[(df3.index >= '2022-06-09 16:02:20') & (df3.index <= '2022-06-09 16:02:43'),['RightTurn']] = 7
df3.loc[(df3.index >= '2022-06-09 16:03:07') & (df3.index <= '2022-06-09 16:03:33'),['RightTurn']] = 7
# Assign left turn based on the log
df3.loc[(df3.index >= '2022-06-09 16:04:26') & (df3.index <= '2022-06-09 16:04:37'),['LeftTurn']] = 5
df3.loc[(df3.index >= '2022-06-09 16:05:08') & (df3.index <= '2022-06-09 16:05:21'),['LeftTurn']] = 5
df3.loc[(df3.index >= '2022-06-09 16:05:40') & (df3.index <= '2022-06-09 16:05:52'),['LeftTurn']] = 5
df3.loc[(df3.index >= '2022-06-09 16:06:36') & (df3.index <= '2022-06-09 16:06:46'),['LeftTurn']] = 5
```

```python
plt.figure(figsize=(12.5, 12.5))
df3.absoluteAltitude.plot(label = "absoluteAltitude")
df3.groundSpeed.plot(label = "groundSpeed")
df3.HoverOrNot.plot(label = "HoverOrNot")
df3.RightTurn.plot(label = "LogRightTurn")
df3.LeftTurn.plot(label = "logLeftTurn")
df3.LabHoverOrNot.plot(label = "LabelHoverOrNot")
df3.LabRightTurn.plot(label = "LabelRightTurn")
df3.LabLeftTurn.plot(label = "LabelLeftTurn")
plt.title('Hover Maneuvers')
plt.ylabel('Absolute Altitude (Feet)')
plt.xlabel('Date Time')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score

hover_true = df3.HoverOrNot
hover_pred = df3.LabHoverOrNot

Hover_acc = accuracy_score(hover_true, hover_pred)
print(Hover_acc)

RTurn_true = df3.RightTurn
RTurn_pred = df3.LabRightTurn

RTurn_acc = accuracy_score(RTurn_true, RTurn_pred)


LTurn_true = df3.LeftTurn
LTurn_pred = df3.LabLeftTurn

LTurn_acc = accuracy_score(LTurn_true, LTurn_pred)

acc = {'Maneuver': ['Hover', 'Right Turn', 'Left Turn'], 'Accuracy': [Hover_acc, RTurn_acc, LTurn_acc]}

df_acc = pd.DataFrame(acc)
df_acc
```
