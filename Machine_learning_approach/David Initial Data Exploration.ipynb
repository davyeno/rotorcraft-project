```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SimData_2022.06.09_11.43.06.csv.roar.csv',
                parse_dates = {'datetime': ['date', 'time']},
                low_memory=False)
```

```python
df
```

```python
# Convert array of integers to pandas series
numbers_series = pd.Series(df.absoluteAltitude)
  
# Get the window of series
# of observations of specified window size
windows = numbers_series.rolling(50)
  
# Create a series of moving
# averages of each window
moving_averages = windows.mean()
  
# Convert pandas series back to list
moving_averages_list = moving_averages.tolist()
  
# Remove null entries from the list
final_list = moving_averages_list[50 - 1:]
  
print(final_list)
```

```python
plt.plot(final_list)
```

```python
# Convert array of integers to pandas series
numbers_series = pd.Series(df.groundSpeed)
  
# Get the window of series
# of observations of specified window size
windows = numbers_series.rolling(50)
  
# Create a series of moving
# averages of each window
moving_averages = windows.mean()
  
# Convert pandas series back to list
moving_averages_list = moving_averages.tolist()
  
# Remove null entries from the list
ground_speed = moving_averages_list[50 - 1:]
  
print(ground_speed)
```

```python
plt.plot(ground_speed)
```

```python
df2 = df[['datetime',
          'latitude',
          'longitude', 
          'absoluteAltitude', 
          'trueHeading', 
          'groundSpeed', 
          'trueAirspeed', 
          'climbOrDescentRate', 
          'hasWeightOnWheels', 
          'turnRate to degPerSec', 
          'cyclicPositionPitch', 
          'cyclicPositionRoll', 
          'collectivePosition', 
          'antiTorquePedalPosition', 
          'throttlePosition', 
          'e1Torque', 
          'e2Torque',
]]
```

```python
df2
```

```python
df2['datetime'] = df2.datetime.dt.strftime('%d/%m/%y %H:%M:%S')
```

```python
df3 = df2.groupby(by=["datetime"], dropna=False).mean().reset_index()
```

```python
plt.figure(figsize=(12.5, 12.5))
plt.plot('absoluteAltitude',
         data = df3,
         label = "absoluteAltitude")
plt.plot('groundSpeed',
         data = df3,
         label = "groundSpeed")
# plt.plot('climbOrDescentRate',
#          data = df3,
#          label = "climbOrDescentRate")
plt.plot('HoverOrNot',
         data = df3,
         label = "HoverOrNot")

plt.legend()
plt.show()
```

```python
df3['HoverOrNot'] = 0
```

```python
for index, row in df3.iterrows():
    if row['absoluteAltitude'] > 2 
    and row['groundSpeed'] < 2 
    and row['climbOrDescentRate'] >= -90 and row['climbOrDescentRate'] <= 90:
        df3['HoverOrNot'][index] = 10
```
